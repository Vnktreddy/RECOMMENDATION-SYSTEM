```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data=pd.read_csv('C:/Users/Admin/Downloads/book.csv')
data
```

```python
data['User.ID']
```

```python
data=data.drop('Unnamed: 0',axis=1)
data
```

```python
data = data.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating')
data
```

```python
data.fillna(0,inplace=True)
data
```

```python
data_dist = 1-pairwise_distances(data.values, metric = 'cosine')
np.fill_diagonal(data_dist,0)
data_dist = pd.DataFrame(data_dist)
data_dist
```

```python
data_dist.index = data['User.ID'].unique()
data_dist.columns = data['User.ID'].unique()
data_dist
```
